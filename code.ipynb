```python
import pandas as pd
french = dict()
df = pd.read_csv('/content/french_dictionary.csv')
french['abide'] = 'respecter'
for i in range(len(df)):
  french[df.iloc[i,0]] = df.iloc[i,1]
```

```python
import re
s = set()
d = dict()

ans = open('output.txt','w')

f2 = open('/content/find_words.txt', 'r')
lines = f2.readlines()
words = []
for i in range(len(lines)):
  words.extend(lines[i].split())
print(words)

file = open('/content/t8.shakespeare.txt','r')
f = file.readlines()
for i in range(len(f)):
  # ans.write(f[i])
  for word in f[i].split(' '):

    temp = ""
    wordlen = len(word)

    for i in word:
      if i.isalpha():
        temp += i
    templen = len(temp)
    lower = temp.lower()
    if lower in words:
      towrite = ""
      if(temp.isupper()):
        towrite = french[lower].upper()
      elif(temp[0].isupper()):
        towrite = french[lower][0].upper() + french[lower][1:]
      else:
        towrite = french[lower]
      
      if(wordlen == templen):
        ans.write(towrite)
      else:
        diff = wordlen - templen
        sl = slice(0, diff)
        if(word[sl] == temp[sl]):
          ans.write(towrite + word[wordlen-diff:])
        else:
          ans.write(word[sl] + towrite)

      s.add(lower)
      if lower in d.keys():
        d[lower] += 1
      else:
        d[lower] = 1
    else:
      ans.write(word)
    ans.write(' ')

```

```python
import csv
field_names = ['English Word', 'French Word', 'Frequency']
dict_data = []
for key in d.keys():
   t = {
       'English Word': key,
       'French Word': french[key],
       'Frequency': d[key]
   }
   dict_data.append(t)
  
with open('frequency.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    writer.writerows(dict_data)
```
